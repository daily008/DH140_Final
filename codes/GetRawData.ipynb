```python
from elsapy.elsclient import ElsClient
from elsapy.elsprofile import ElsAuthor, ElsAffil
from elsapy.elsdoc import FullDoc, AbsDoc
from elsapy.elssearch import ElsSearch
import json
import re
import csv
import xlwt



def read_json(input):
    author = []
    all_author = input["coredata"]['dc:creator']

    for v in all_author:
        try:
            author.append(v['$'])
        except TypeError as e:
            print(e)
            print(v)
            return [all_author['$']]
        return author

def get_pii(url): #根据doi获得pii
    value = re.search("pii/S.*",url).group()
    return value[4:]

def get_doi(url): #根据doi获得10.XX.XX
    value = re.search("org/10.*",url).group()
    return value[4:]


if __name__ == "__main__":
    ## Load configuration
    con_file = open("config.json")
    config = json.load(con_file)
    con_file.close()

    ## Initialize client
    client = ElsClient(config['apikey'])

    all_datapaper_doi = []
    all_originalpaper_doi = []
    with open("output/RawData.csv", encoding="utf-8") as f:
    #with open("data/data_in_brief/data_in_brief_doi.csv",encoding= "utf-8") as f:
        for row in csv.reader(f, skipinitialspace=True):
            all_datapaper_doi.append(row[0]) #收集doi信息
            all_originalpaper_doi.append(row[1])
        f.close()
    all_datapaper_doi = all_datapaper_doi[1:] #这一部处理是为了去掉首行的label
    all_originalpaper_doi = all_originalpaper_doi[1:]

    file = open("output/data_paper_author_output.csv", 'a+', encoding= "utf-8-sig", newline='')
    csv_writer = csv.writer(file)

    for item in all_originalpaper_doi:
        rank = 0
        print ("12")
        if item == "":
            print("222")
            break

        pii = get_pii(item)
        pii_doc = FullDoc(sd_pii=pii)

        if pii_doc.read(client):
            all_data = pii_doc.data
            document_doi = pii_doc.uri
            pii_doc.write()
            authors = read_json(all_data) #从返回的json格式的bibinformation中，提取作者信息，格式为列表 【author1,author2,...】，姓在前名在后

            for person in authors:
                rank = rank + 1
                csv_writer.writerow([item, document_doi, person, rank])
        else:
            print ("Read document failed.")
            csv_writer.writerow([item,"failed","failed"])

    for item in all_datapaper_doi:
        rank = 0
        if item is None:
            break
        k = get_doi(item)
        pii_doc = FullDoc(doi= k)

        if pii_doc.read(client):
            all_data = pii_doc.data
            document_doi = pii_doc.uri
            pii_doc.write()
            authors = read_json(all_data) #从返回的json格式的bibinformation中，提取作者信息，格式为列表 【author1,author2,...】，姓在前名在后
            for person in authors:
                rank = rank + 1

                csv_writer.writerow([item, document_doi, person, rank])
        else:
            print ("Read document failed.")
            csv_writer.writerow([item,"failed","failed"])

    file.close()
```
